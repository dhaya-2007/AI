```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Step 2: Create dataset
data = {
    'Score': [30, 35, 40, 45, 50, 55, 60, 65],
    'Result': [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# Step 3: Define X and y
X = df[['Score']]   # 2D input
y = df['Result']

# Step 4: Create model
model = LogisticRegression()

# Step 5: Train model
model.fit(X, y)

# Step 6: Predict
df['Predicted'] = model.predict(X)

# Step 7: Accuracy
print("Accuracy:", model.score(X, y))

# Step 8: Plot
plt.scatter(df['Score'], df['Result'], label='Actual')
plt.scatter(df['Score'], df['Predicted'], marker='x', label='Predicted')
plt.xlabel('Exam Score')
plt.ylabel('Result (0=Fail, 1=Pass)')
plt.legend()
plt.show()
```

```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Step 2: Create dataset
data = {
    'Score': [30, 35, 40, 45, 50, 55, 60, 65],
    'Result': [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# Step 3: Define X and y
X = df[['Score']]   # 2D input
y = df['Result']

# Step 4: Create model
model = LogisticRegression()

# Step 5: Train model
model.fit(X, y)

# Step 6: Predict
df['Predicted'] = model.predict(X)

# Step 7: Accuracy
print("Accuracy:", model.score(X, y))

# Step 8: Plot
plt.scatter(df['Score'], df['Result'], label='Actual')
plt.scatter(df['Score'], df['Predicted'], marker='x', label='Predicted')
plt.xlabel('Exam Score')
plt.ylabel('Result (0=Fail, 1=Pass)')
plt.legend()
plt.show()

# Step 9: Calculate and print Confusion Matrix
cm = confusion_matrix(y, df['Predicted'])
print("\nConfusion Matrix:")
print(cm)
```
